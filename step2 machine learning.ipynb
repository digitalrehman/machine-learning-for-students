## **There are three main types of machine learning**


- **Supervised Learning**: Jab data ke sath labels diye hote hain (input + output). Model ko sikhaya jata hai ke input se output predict kare.
`Example`: Email spam detection.
- **Unsupervised Learning**: Jab data ke sath labels nahi hote. Model khud patterns ya groups find karta hai.
`Example`: Customer segmentation.
- **Reinforcement Learning**: Jab ek agent environment main actions leta hai aur reward/punishment ke zariye seekhta hai.
`Example`: Game AI jo har move ke baad reward leta hai.


*Encoding ek data preprocessing step hai jo mostly supervised learning main use hota hai. Unsupervised learning main kabhi kabhi use hota hai (agar categorical data ho), aur reinforcement learning main kam zarurat hoti hai.*
