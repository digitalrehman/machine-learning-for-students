## **There are three main types of machine learning**


- **Supervised Learning**: Jab data ke sath labels diye hote hain (input + output). Model ko sikhaya jata hai ke input se output predict kare.

`Example`: Email spam detection.
- **Unsupervised Learning**: Jab data ke sath labels nahi hote. Model khud patterns ya groups find karta hai.

`Example`: Customer segmentation.
- **Reinforcement Learning**: Jab ek agent environment main actions leta hai aur reward/punishment ke zariye seekhta hai.

`Example`: Game AI jo har move ke baad reward leta hai.


*Encoding ek data preprocessing step hai jo mostly supervised learning main use hota hai. Unsupervised learning main kabhi kabhi use hota hai (agar categorical data ho), aur reinforcement learning main kam zarurat hoti hai.*

___

### **MinMaxScaler**
- Har feature ko ek fixed range (default: 0 to 1) main scale karta hai.
- Useful jab features ke values different scales pe ho aur aapko uniform range chahiye.



```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Sample data
df = pd.DataFrame({
    'Age': [20, 25, 30, 35, 40],
    'Salary': [20000, 25000, 30000, 35000, 40000]
})

# MinMaxScaler (0 to 1 range)
minmax = MinMaxScaler()
pd.DataFrame(minmax.fit_transform(df), columns=df.columns)
```

### **StandardScaler**
- Har feature ko mean = 0 aur standard deviation = 1 ke scale pe normalize karta hai.
- Useful jab data normal distribution ke around ho aur ML algorithms (jaise Logistic Regression, SVM) sensitive ho scaling pe.

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler (mean=0, std=1)
standard = StandardScaler()
pd.DataFrame(standard.fit_transform(df), columns=df.columns)
```

### ***Train Test Split***

```python
from sklearn.model_selection import train_test_split

# Features (X) and Target (y)
X = df[["Age"]]  # independent variable
y = df["Salary"]  # dependent variable

# Split dataset (train 70%, test 30%)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
print("X_train:\n", X_train)
print("y_train:\n", y_train)

print("X_test:\n", X_test)
print("y_test:\n", y_test)
```

#### **Independent Variable (Feature, X)**
- Ye wo input hota hai jo model ko dete ho.
- Isko predictor variable bhi kehte hain.
- `Example`: Agar tum Age aur Experience se Salary predict karna chahte ho, to Age aur Experience independent variables hain.


#### **Dependent Variable (Target, y)**
- Ye wo output hota hai jo model predict karta hai.
- Isko response variable bhi kehte hain.
- `Example`: Salary dependent variable hai, kyunki wo Age aur Experience pe depend karti hai.


#### **Short Example**

##### Age aur Experience -> independent variables (X)
##### Salary -> dependent variable (y)

X = df[['Age', 'Experience']]   # independent variables

y = df['Salary']                # dependent variable
